```python
#1.	Create the below pattern using nested for loop in Python.
'''   
*
** 
** * 
** * * 
** * * * 
** * * 
** * 
** 
*'''

n=int(input("Enter the length "))
for i in range(n-1):
    for j in range(i+1):
        if j<1:
            print("*",end="")
        else:
            print("*",end=" ")
    print()  

for k in range(n,0,-1):
    for l in range(k):
        if <1:
            print("*",end="")
        else:
            print("*",end=" ")
    print()    
```
